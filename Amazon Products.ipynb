```python
#Category model
category = [ 
{'Id' :1 , 'Category': 'Electronics'},
    {'Id' :2 , 'Category': 'Media'},
    {'Id' :3 , 'Category': 'Miscellaneous'},
    {'Id' :4 , 'Category': 'Home'},
    {'Id' :5 , 'Category': 'fashion'},
    {'Id' :6 , 'Category': 'Food'},
]

category_df = pd.DataFrame(category)
category_df
```

```python
#Joining under category name

import sqlite3
conn = sqlite3.connect(':memory:')
products_df.to_sql('products', conn, if_exists = 'replace')
category_df.to_sql('category', conn, if_exists = 'replace')

query = """

select p.* , c.category as CategName from products p left join category c 
on p.category =  c.id

"""

union = pd.read_sql(query, con = conn)
union

```

```python
#Popularity by category
unions = union[['CategName', 'SearchRate(Million)']].groupby('CategName').aggregate('sum')
unions
```
