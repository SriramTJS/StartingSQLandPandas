```python
#Connecting to SQL query for inner join

import sqlite3
conn = sqlite3.connect(':memory:')
orders_df.to_sql("orders", conn, if_exists="replace")
cust_dfs.to_sql("cust", conn, if_exists="replace")

query = """
select o.*, c.name from orders o inner join cust c on c.customer_id = o.customer_id order by o.customer_id
"""
result_df = pd.read_sql(query, con=conn)
result_df
```

```python
#Total sales by name

results = result_df.groupby([ 'name', 'customer_id']).aggregate('sum')
results
```
